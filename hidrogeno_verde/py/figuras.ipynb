#Gr√°fica 1. Barras verticales

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Lee el archivo CSV en un DataFrame
df = pd.read_csv('electricity-renewables.csv',sep=',')
pais=df["Entity"].value_counts()
electricity=df["Electricity from renewables - TWh"].value_counts()
df=df[df["Electricity from renewables - TWh"]>30]

#Gr√°fica 1
eje_x = df["Entity"]
eje_y = df["Electricity from renewables - TWh"]
fig, axs = plt.subplots(1,1)

axs.bar(eje_x, eje_y, color='blue', edgecolor='black')
rects = axs.bar(eje_x, eje_y)
axs.bar_label(rects, fmt='%.1f', label_type='edge', fontsize=6, rotation=45)

axs.set_box_aspect(0.6)

plt.xticks(rotation=60, fontsize=8)
plt.title("Energ√≠as Renovables en el Mundo entre 2023 y 2024")
plt.xlabel('Pa√≠ses', fontsize=10)
plt.ylabel('Producci√≥n energ√≠a renovable en TW-h')

plt.show()

```

Gr√°fica 2. Barras horizontales

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Datos 
regiones = ['2014-2023', 'STEPS2024-2035', 'STEPS2036-2050', 'APS2024-2035', 'APS2036-2050']  	

Biometano =[1, 5, 10, 15, 32]
Biogas = [1, 5, 7, 8, 12]
GN = [0, 4, 3, 4, 2]
Hidrogeno = [0, 8, 13, 37, 46]

fig, ax = plt.subplots()

y = np.arange(len(regiones))

ax.barh(y, Biometano, label='2014-2023', color='skyblue', edgecolor= 'black')
ax.barh(y, Biogas, left=Biometano, label='STEPS2024-2035', color='lightgreen', edgecolor= 'black')
ax.barh(y, GN, left=np.array(Biometano) + np.array(Biogas), label='STEPS2036-2050', color='cyan', edgecolor= 'black')
ax.barh(y, Hidrogeno, left=np.array(Biometano) + np.array(Biogas) + np.array(GN), label='APS2024-2035', color='violet', edgecolor= 'black')

ax.set_xlabel('billones de US')
ax.set_ylabel('Escenarios de producci√≥n de "Low-Emissions"')
ax.set_title('Inversi√≥n promedio anual de tecnolog√≠as en bajas emisiones 2014 - 2050')
ax.set_yticks(y)
ax.set_yticklabels(regiones)
ax.legend()

plt.show()
```
